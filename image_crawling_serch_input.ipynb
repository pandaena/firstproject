```python
# burn inhury

# 경로확인
# d드라이브에 이미지 저정할 예정이라 d드라이브로 디렉토리 이동

%pwd

# selenium  설치
!pip install selenium

# 패키지 설치
from selenium import webdriver
from selenium.webdriver.common.by import By
from urllib.request import urlopen
from selenium.webdriver.common.keys import Keys
import time
import urllib.request
import os

# 검색어 지정
search = str(input("검색어를 지정하세요 : "))


# 이미지 저장 폴더 만들기
if not os.path.isdir(search + "\\"):
    os.makedirs(search + "\\")

# 이용할 드라이버 Chrome으로 지정
driver = webdriver.Chrome()
driver.get("https://www.google.co.kr/imghp?hl=ko&ogbl")

# 원격 서버가 헤더를 확인하고 요청 거부할 때 생기는 urllib.error.HTTPError: HTTP Error 403: Forbidden 에러 방지
opener = urllib.request.build_opener()
opener.addheaders = [('User-Agent', 'Mozila/5.0')]
urllib.request.install_opener(opener)


# 검색창 찾기 ("q"를 넣어주는 것은 검색창을 찾는 것) 
# 검색창을 찾고 search 키워드를 넣음
# 엔터를 쳐줌
elem = driver.find_element(By.NAME, "q")
elem.send_keys(search)
elem.send_keys(Keys.RETURN)

# 연관검색을 위해 시간 벌기
time.sleep(20)

# # 연관 검색을 위해 이미지 썸네일을 찾음
# # 안 바꿔줘도 됨
# images = driver.find_elements(By.CSS_SELECTOR, ".rg_i.Q4LuWd") 

# 화면 내리기
SCROLL_PAUSE_TIME = 1

# 구글 이미지 검색이 스크롤을 내리면 계속 새로운 스크롤이 갱신된다는 특성을 이용
# 구글 이미지 검색 시 scrollHeight이 끝나는 지점에 다다랐을 때 "결과 더보기" 버튼이 생성됨
# 버튼을 찾으면 누르고, 아닐 경우 스크롤 내리기를 멈춤
# 안 바꿔줘도 됨
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(SCROLL_PAUSE_TIME)
    new_height = driver.execute_script("return document.body.scrollHeight")

    if new_height == last_height:
        try:
            driver.find_element(By.CSS_SELECTOR,".mye4qd").click() #더보기 버튼
        except:
            break
    last_height = new_height


# 이미지 썸네일 찾음
# 안 바꿔줘도 됨
images = driver.find_elements(By.CSS_SELECTOR, ".rg_i.Q4LuWd") 
count = 1


# 이미지 클릭 후 xpath에 붙어있는 src(소스) 정보를 가져옴
# urllib.request.urlretrieve 통해서 이미지 저장
# try를 이용해서 오류가 생기는 이미지 패스
# 안 바꿔줘도 됨
for image in images:
    try:
        image.click()
        time.sleep(2)


        imgUrl = driver.find_element(By.XPATH, "//*[@id='Sva75c']/div[2]/div/div[2]/div[2]/div[2]/c-wiz/div[2]/div[1]/div[1]/div[2]/div/a/img").get_attribute('src')
        urllib.request.urlretrieve(imgUrl, "C:\\Users\\admin\\" +search + "\\" + search + "_" + str(count) + ".jpg")
        print("Image saved: " + search + "_{}.jpg".format(count))
        count += 1
    except Exception as e:
        print(str(e))
        pass

driver.close()
```
